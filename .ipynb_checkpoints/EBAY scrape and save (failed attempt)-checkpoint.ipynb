```python
def scrape_product_link(link):
    driver = webdriver.Chrome()
    driver.get(link)
    driver.implicitly_wait(5)
    products = pd.DataFrame(data={
        'id':[],
        'title':[],
        'price':[],
        'description':[],
        'link':[],
        'img_url':[],
        'condition':[],
        'positive_reviews':[],
        'neutral_reviews':[],
        'negative_reviews':[],
    })
    
    page = 20

    while(page != 0):
        # Find all product links
        product_links = driver.find_elements(By.CSS_SELECTOR, 'a.s-item__link')
        
        # Iterate over each product link
        for product_link in product_links:
            try:
                # Scroll element into view
                driver.execute_script("arguments[0].scrollIntoView();", product_link)
                
                # Wait for the element to be clickable
                WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'a.s-item__link')))
                
                # Click on the element
                product_link.click()
                
                # Switch to the newly opened tab
                driver.switch_to.window(driver.window_handles[-1])
                time.sleep(4)  # Adjust as necessary
                print('**************************************')
                try: # done as some values are not on the website 
                    Id = driver.find_element(By.CSS_SELECTOR, 'div.tabs__content div.ux-layout-section__textual-display.ux-layout-section__textual-display--itemId span.ux-textspans.ux-textspans--BOLD').text.strip()
                except:
                    continue
                try:
                    title = driver.find_element(By.CSS_SELECTOR, 'h1.x-item-title__mainTitle span.ux-textspans.ux-textspans--BOLD').text.strip()
                except:
                    continue            
                try:
                    price = driver.find_element(By.CSS_SELECTOR, 'div.x-price-primary span.ux-textspans').text.strip()
                except:
                    continue 
                try:
                    condition = driver.find_element(By.CSS_SELECTOR, 'div.x-item-condition-text span.ux-textspans').text.strip()
                except:
                    continue 
                try:
                    img_url = driver.find_element(By.CSS_SELECTOR, 'div.ux-image-carousel-item.image-treatment.active.image img').get_attribute('src')
                except:
                    continue 
                try: 
                    description = []
                    description_rows = driver.find_element(By.CSS_SELECTOR,'div[class="ux-layout-section-evo ux-layout-section--features"]').find_elements(By.CSS_SELECTOR, 'div.ux-layout-section-evo__row')
                    for row in description_rows:
                        cols = row.find_elements(By.CSS_SELECTOR, 'div.ux-layout-section-evo__col')
                        for col in cols:
                            # the label
                            try:
                                feature_label = col.find_element(By.CSS_SELECTOR, 'div.ux-labels-values__labels-content').find_element(By.CSS_SELECTOR,'span[class="ux-textspans"]').text.strip()
                            except:
                                continue
                            # the description
                            try:
                                feature_content = col.find_element(By.CSS_SELECTOR, 'div.ux-labels-values__values-content span.ux-textspans').text.strip()
                            except NoSuchElementException:
                                feature_content = col.find_element(By.CSS_SELECTOR, 'div.ux-labels-values__values-content').find_element(By.CSS_SELECTOR,'span[class="ux-textspans ux-textspans--BOLD"]').text.strip()
                            description.append(f"{feature_label} - {feature_content}")
                except:
                    continue 
        
        
        
                # Example: Navigate to reviews section and scrape reviews
                print('**************************************')
                try:
                    print('**************************************try')    
                    try:
                        review_link = feedback_tbody = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, 'div[class="fdbk-detail-list__btn-container"] a'))
                    )
                    except:
                        review_link = WebDriverWait(driver, 10).until(
                            EC.presence_of_element_located((By.CSS_SELECTOR, 'a.fdbk-detail-list__tabbed-btn.fake-btn.fake-btn--large.fake-btn--secondary'))
                        )
                        # Scroll into view and click the feedback button
                    driver.execute_script("arguments[0].scrollIntoView(true);", review_link)
                    print('**************************************review_link')
                    # fetched the a tag element from the code. This way i can use the click operation
                    review_link.click()
                    print('**************************************')
                    feedback_tbody = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, 'table[id="feedback-cards"] tbody'))
                    )
                    print('**************************************positive')
                    positive_feedback_button = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR,  'button[data-test-id="rating-count-3"]'))
                    )
                    positive_feedback_button.click()
                    #Operation of scrape
                    positive_feedback=[]
                    feedback_table = driver.find_element(By.CSS_SELECTOR, 'table[id="feedback-cards"] tbody').find_elements(By.TAG_NAME,'tr')
                    for card_container in feedback_table:
                        try:  # done as there are replies within the feedbacks
                            review = card_container.find_element(By.CSS_SELECTOR,'div[class="card__comment"] span').text.strip() 
                        except: 
                            continue
                        positive_feedback.append(review)
                    print('**************************************neutral')
                    neutral_feedback_button = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR,  'button[data-test-id="rating-count-6"]'))
                    )
                    neutral_feedback_button.click()
                    #operarion of scraping
                    neutral_feedback=[]
                    feedback_table = driver.find_element(By.CSS_SELECTOR, 'table[id="feedback-cards"] tbody').find_elements(By.TAG_NAME,'tr')
                    for card_container in feedback_table:
                        try:
                            review = card_container.find_element(By.CSS_SELECTOR,'div[class="card__comment"] span').text.strip()
                        except:
                            continue
                        neutral_feedback.append(review) 
                    print('**************************************negative')
                    negative_feedback_button = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, 'button[data-test-id="rating-count-9"]'))
                    )
                    negative_feedback_button.click()
                    # scraping operation
                    negative_feedback=[]
                    feedback_table = driver.find_element(By.CSS_SELECTOR, 'table[id="feedback-cards"] tbody').find_elements(By.TAG_NAME,'tr')
                    for card_container in feedback_table:
                        try:
                            review = card_container.find_element(By.CSS_SELECTOR,'div[class="card__comment"] span').text.strip()
                        except:
                            continue
                        negative_feedback.append(review)
                        print(negative_feedback)
                except:
                    continue
        
                
                # After scraping, go back to the original product page
                driver.back()
        
                product = pd.DataFrame(data={
                    'id': [Id],
                    'title': [title],
                    'price': [price],
                    'description': [description],
                    'link': [product_link.get_attribute('href')],
                    'img_url': [img_url],
                    'condition': [condition],
                    'positive_reviews': [positive_feedback],
                    'neutral_reviews': [neutral_feedback],
                    'negative_reviews': [negative_feedback],
                })
                print('sql query')
                
                cursor.execute('SELECT * FROM products')
                records = cursor.fetchall()
                for record in records:
                    print(record)

                
                print('a') 
                driver.close()
                print('b')
                # Switch back to the main tab
                driver.switch_to.window(driver.window_handles[0])
                print('c')
                #products = pd.concat([products,product],axis=0,ignore_index=True)  
                print('d')
                print(products)
                print('e')
            except Exception as e:
                print(f"Exception occurred ---: {str(e)}")
                driver.switch_to.window(driver.window_handles[0])
                continue  # Skip to the next product link if there's an issue
    
        # Example: Click the "Next" button for pagination (repeat as necessary)
        # Replace with actual pagination handling logic
        next_button = driver.find_element(By.CSS_SELECTOR, 'nav[class="pagination"]').find_element(By.CSS_SELECTOR, 'a[class="pagination__next icon-link"]')
        next_button.click()
        page-=1
        time.sleep(3)
    return products
    driver.quit()
```

```python
import psycopg2
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.get(r"https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2555059.m570.l1313&_nkw=cameras&_sacat=0")
driver.implicitly_wait(5)

# Initialize the DataFrame
products = pd.DataFrame(data={
    'id': [],
    'title': [],
    'price': [],
    'description': [],
    'link': [],
    'img_url': [],
    'condition': [],
    'positive_reviews': [],
    'neutral_reviews': [],
    'negative_reviews': [],
})

# Fetch product links
product_links = driver.find_elements(By.CSS_SELECTOR, 'a.s-item__link')

# Iterate over each product link
for product_link in product_links:
    try:
        # Scroll element into view
        driver.execute_script("arguments[0].scrollIntoView();", product_link)
        
        # Wait for the element to be clickable
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'a.s-item__link')))
        
        # Click on the element
        product_link.click()
        
        # Switch to the newly opened tab
        driver.switch_to.window(driver.window_handles[-1])
        time.sleep(4)  # Adjust as necessary
        try:
            Id = driver.find_element(By.CSS_SELECTOR, 'div.tabs__content div.ux-layout-section__textual-display.ux-layout-section__textual-display--itemId span.ux-textspans.ux-textspans--BOLD').text.strip()
        except NoSuchElementException:
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            continue

        try:
            title = driver.find_element(By.CSS_SELECTOR, 'h1.x-item-title__mainTitle span.ux-textspans.ux-textspans--BOLD').text.strip()
        except NoSuchElementException:
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            continue
            
        try:
            price = driver.find_element(By.CSS_SELECTOR, 'div.x-price-primary span.ux-textspans').text.strip()
        except NoSuchElementException:
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            continue
            
        try:
            condition = driver.find_element(By.CSS_SELECTOR, 'div.x-item-condition-text span.ux-textspans').text.strip()
        except NoSuchElementException:
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            continue
            
        try:
            img_url = driver.find_element(By.CSS_SELECTOR, 'div.ux-image-carousel-item.image-treatment.active.image img').get_attribute('src')
        except NoSuchElementException:
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            continue
            
        try: 
            description = []
            description_rows = driver.find_element(By.CSS_SELECTOR, 'div[class="ux-layout-section-evo ux-layout-section--features"]').find_elements(By.CSS_SELECTOR, 'div.ux-layout-section-evo__row')
            for row in description_rows:
                cols = row.find_elements(By.CSS_SELECTOR, 'div.ux-layout-section-evo__col')
                for col in cols:
                    try:
                        feature_label = col.find_element(By.CSS_SELECTOR, 'div.ux-labels-values__labels-content').find_element(By.CSS_SELECTOR, 'span[class="ux-textspans"]').text.strip()
                    except NoSuchElementException:
                        continue
                    try:
                        feature_content = col.find_element(By.CSS_SELECTOR, 'div.ux-labels-values__values-content span.ux-textspans').text.strip()
                    except NoSuchElementException:
                        feature_content = col.find_element(By.CSS_SELECTOR, 'div.ux-labels-values__values-content').find_element(By.CSS_SELECTOR, 'span[class="ux-textspans ux-textspans--BOLD"]').text.strip()
                    description.append(f"{feature_label} - {feature_content}")
        except NoSuchElementException:
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
            continue

        positive_feedback = []
        neutral_feedback = []
        negative_feedback = []

        try:
            try:
                review_link = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, 'div[class="fdbk-detail-list__btn-container"] a'))
                )
            except:
                review_link = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, 'a.fdbk-detail-list__tabbed-btn.fake-btn.fake-btn--large.fake-btn--secondary'))
                )
            driver.execute_script("arguments[0].scrollIntoView(true);", review_link)
            review_link.click()

            feedback_tbody = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'table[id="feedback-cards"] tbody'))
            )

            positive_feedback_button = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'button[data-test-id="rating-count-3"]'))
            )
            positive_feedback_button.click()
            feedback_table = driver.find_element(By.CSS_SELECTOR, 'table[id="feedback-cards"] tbody').find_elements(By.TAG_NAME, 'tr')
            for card_container in feedback_table:
                try:
                    review = card_container.find_element(By.CSS_SELECTOR, 'div[class="card__comment"] span').text.strip()
                    positive_feedback.append(review)
                except NoSuchElementException:
                    continue
            print('extration of positive reviews successful')
            
            neutral_feedback_button = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'button[data-test-id="rating-count-6"]'))
            )
            neutral_feedback_button.click()
            feedback_table = driver.find_element(By.CSS_SELECTOR, 'table[id="feedback-cards"] tbody').find_elements(By.TAG_NAME, 'tr')
            for card_container in feedback_table:
                try:
                    review = card_container.find_element(By.CSS_SELECTOR, 'div[class="card__comment"] span').text.strip()
                    neutral_feedback.append(review)
                except NoSuchElementException:
                    continue
            print('extration of neutral reviews successful')
            
            negative_feedback_button = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'button[data-test-id="rating-count-9"]'))
            )
            negative_feedback_button.click()
            feedback_table = driver.find_element(By.CSS_SELECTOR, 'table[id="feedback-cards"] tbody').find_elements(By.TAG_NAME, 'tr')
            for card_container in feedback_table:
                try:
                    review = card_container.find_element(By.CSS_SELECTOR, 'div[class="card__comment"] span').text.strip()
                    negative_feedback.append(review)
                except NoSuchElementException:
                    continue
            print('extration of negative reviews successful')
            
        except TimeoutException:
            pass
        except NoSuchElementException:
            pass
        except Exception as e:
            pass

        # Insert data into PostgreSQL
        try:
            insert_query = '''
            INSERT INTO products (id, title, price, description, link, img_url, condition, positive_reviews, neutral_reviews, negative_reviews)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
            ON CONFLICT (id) DO NOTHING;
            '''
            cursor.execute(insert_query, (
                Id, title, price, description, product_link.get_attribute('href'),
                img_url, condition, positive_feedback, neutral_feedback, negative_feedback
            ))
            conn.commit()
            print('sql query successful')
        except Exception as e:
            pass

        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        print('*************************************************** next')
    except Exception as e:
        driver.switch_to.window(driver.window_handles[0])
        print('**************-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*next with exception')
```
